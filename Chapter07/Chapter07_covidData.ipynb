### Covid19 API: Python - Data


```python
import requests
import csv
```

```python
baseUrl="https://covid19api.com/"
apiUrl="https://api.covid19api.com/summary"
```

```python
response = requests.get(apiUrl).json()
```

#### Total Countries Data 

```python
len(response['Countries']) 
```

#### Collecting only fields 'Country','TotalConfirmed','TotalDeaths' 

```python
lists=[]
```

```python
for info in response['Countries']:
    percentage=(info['TotalDeaths']/info['TotalConfirmed'])*100
    lists.append([
            info['Country'],info['TotalConfirmed'],info['TotalDeaths'],percentage,round(percentage)
        ])    
```

```python
lists
```

#### Convert List(dataSet) to CSV

```python
def writeto_csv(data,filename,columns):
    with open(filename,'w+',newline='',encoding="UTF-8") as file:
        writer = csv.DictWriter(file,fieldnames=columns) #Column header
        writer.writeheader()
        writer = csv.writer(file)
        for element in data:
            writer.writerows([element])
```

```python
writeto_csv(lists,'covid_data.csv',['country','total_confirmed','total_death','death_per','death_per_round'])
```
